```python
def get_doc_directory_by_number(dir_list, number):
    for directory in dir_list:
        for item in dir_list[directory]:
            if number in item:
                return(f"Документ хранится на полке: {directory}")
#             print(item)
    return('Документ не найден в базе')

# get_doc_directory_by_number(directories, input())
```

```python
def add_new_directory(dir_list, number):
#     dir_numbers = [i for i in dir_list]
    if number not in dir_list:
#         print('not in list')
        dir_list[number] = []
        return(f"Полка добавлена. Текущий перечень полок: {', '.join(dir_list)}")
    return(f"Такая полка уже существует. Текущий перечень полок: {', '.join(dir_list)}")

# add_new_directory(directories, '98')    
```

Пункт 5. Пользователь по команде "ds" может удалить существующую полку из данных (только если она пустая)
Примеры работы:

1.

Введите команду:
ds

Введите номер полки:
3
Результат:
Полка удалена. Текущий перечень полок: 1, 2.

2.

Введите команду:
ds

Введите номер полки:
1
Результат:
На полке есть документа, удалите их перед удалением полки. Текущий перечень полок: 1, 2, 3.

3.

Введите команду:
ds

Введите номер полки:
4
Результат:
Такой полки не существует. Текущий перечень полок: 1, 2, 3.

```python
def delete_directory_by_number(dir_list, number):
    if number in dir_list:
        if len(dir_list[number]) > 0:
            return(f"На полке есть документы, удалите их перед удалением полки. Текущий перечень полок: {', '.join(dir_list)}")
        dir_list.pop(number)
        return(f"Полка удалена. Текущий перечень полок: {', '.join(dir_list)}")
    
    return(f"Такой полки не существует. Текущий перечень полок: {', '.join(dir_list)}")
    
# delete_directory_by_number(directories)
```

```python
def main():
    while True:
        user_input = input('Введите команду:\n')
        if user_input == 'p':
            print(get_doc_owner_by_number(documents, input('Введите номер документа:\n')))
        elif user_input == 's':   
            print(get_doc_directory_by_number(directories, input('Введите номер документа:\n')))
        elif user_input == 'l':   
            get_doc_list(documents, directories)
        elif user_input == 'ads':
            print(add_new_directory(directories, input('Введите номер полки:\n')))
        elif user_input == 'ds':
            print(delete_directory_by_number(directories, input('Введите номер полки:\n')))
        elif user_input == 'q':
            break
main()    
```
