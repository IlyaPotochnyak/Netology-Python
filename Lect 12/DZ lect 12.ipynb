По данным файла power.csv посчитайте суммарное потребление стран Прибалтики (Латвия, Литва и Эстония)
категорий 4, 12 и 21 за период с 2005 по 2010 года. Не учитывайте в расчетах отрицательные значения quantity.

```python
power_data = pd.read_csv('power.csv')
power_data.head()
```

```python
power_data[power_data['country'].str.contains('lat', case=False)]['country'].unique()
```

```python
power_data[power_data['country'].str.contains('lit', case=False)]['country'].unique()
```

```python
power_data[power_data['country'].str.contains('est', case=False)]['country'].unique()
```

```python
filtered_countries = power_data[ (power_data['country']=='Latvia') | (power_data['country']=='Lithuania')  | 
                                (power_data['country']=='Estonia')]
filtered_countries.head(20)
```

```python
# filtered_countries = filtered_countries[filtered_countries[(filtered_countries['category']==4) | 
#                                                            (filtered_countries['category']==12)]]

filtered_countries = filtered_countries[(filtered_countries['category'] == 4) | (filtered_countries['category'] == 12) | 
                   (filtered_countries['category'] == 21)]
filtered_countries.head()
```

```python
# filtered_countries = filtered_countries[(filtered_countries['quantity'] >= 0) & 
#                                         (filtered_countries['year'] >= 2005)& 
#                                         (filtered_countries['year'] <= 2010)]

filtered_countries = filtered_countries[(filtered_countries['quantity'] >= 0) & 
                                        (filtered_countries['year'] >= 2005)& 
                                        (filtered_countries['year'] <= 2010) ]

filtered_countries.head(20)
```

```python
filtered_countries['quantity'].sum()
```
